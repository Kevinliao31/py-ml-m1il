```python
import pandas as pd

# Charger les données du fichier CSV
games = pd.read_csv('./db/vgsales.csv')

# 1. Requête qui retourne tous les jeux sortis sur NES
nes_games = games[games['Platform'] == 'NES']
print("Jeux sur NES:")
print(nes_games)

# 2. Requête qui retourne la somme des ventes japonaises de tous les jeux
currency = "€"
JP_sum_sales = round(games['JP_Sales'].sum(), 2)
print("La somme des ventes japonaises s'élèvent à : ",JP_sum_sales,currency)

# 3. Requête qui retourne le nombre de jeux sortis à partir de 2008 
count_released_games_since_2008 = 0

for game in games.itertuples():
    if game.Year >= 2008:
        count_released_games_since_2008 += 1

print("Nombre de jeux sortis à partir de 2008 : ", count_released_games_since_2008)

# 4. Requête suir retourne le pourcentage de jeux Wii
count_wii_games = 0
total_games_count = games.shape[0]

for game in games.itertuples():
    if game.Platform == "Wii":
        count_wii_games += 1
    
wii_percentage = round((count_wii_games / total_games_count) * 100, 2)

print("Pourcentage de jeux Wii : ", wii_percentage, "%")

# 4.1 Autre façon :
wii_games_count = games[games['Platform'] == 'Wii'].shape[0]
print("\nPourcentage de jeux Wii v2 :", wii_percentage, "%")

```
